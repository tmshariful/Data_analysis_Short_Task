```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/task2_p.csv')
df.head()
```

```python
df['Avg_10d_click'] = df.AVERAGE_of_Clicks.rolling(window=10).mean()
df
```

```python
plt.figure(figsize=(12, 8))
plt.title('10-trading days MA')
plt.plot(df.Date, df.Avg_10d_click)
plt.xlabel('Date')
plt.ylabel('AVERAGE_of_Clicks')
plt.xticks(df.Date[::10], rotation='vertical')
plt.legend(('Avg_10d_click'), loc='lower right')
plt.show()
```
